# COVID-19 Clinical Trials EDA

## 1. Loading the Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
try:
    df = pd.read_csv(r'D:\Unified Mentor intern\project 2\COVID clinical trials.csv')
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print("Error: Dataset file not found. Please check the file path.")
```

##  2. Data Exploration

```python
print("First 5 rows of the dataset:")
display(df.head())
print(f"\nThe dataset has {df.shape[0]} rows and {df.shape[1]} columns. [cite: 132]")
print("\nDataset Information:")
df.info()
print("\nDescriptive Statistics:")
display(df.describe(include='all'))
```

## 3. Handling missing data and plotting its Heatmap

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'column_name': df.columns, 'percent_missing': missing_percentage})
missing_df = missing_df[missing_df['percent_missing'] > 0].sort_values('percent_missing', ascending=False)
print("Columns with missing data (%):")
print(missing_df)
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Data')
plt.show()
```

## 4. Dropping Columns With High Missing Values

```python
columns_to_drop = ['Results First Posted', 'Study Documents']
df.drop(columns=columns_to_drop, inplace=True, axis=1)
print(f"Dropped columns: {columns_to_drop}")
```

## 5. Imputing missing values

```python
categorical_cols_with_na = ['Interventions', 'Phases', 'Acronym', 'Gender', 'Study Designs', 'Locations']
for col in categorical_cols_with_na:
    if col in df.columns:
        df[col].fillna('Unknown', inplace=True)
print("Imputed missing values in categorical columns with 'Unknown'.")
enrollment_skew = df['Enrollment'].skew()
print(f"\nSkewness of 'Enrollment' column: {enrollment_skew:.2f} [cite: 455]")
enrollment_median = df['Enrollment'].median()
df['Enrollment'].fillna(enrollment_median, inplace=True)
print(f"Imputed missing 'Enrollment' values with the median ({enrollment_median}).")
df.dropna(inplace=True)
print("\nFinal check for null values:", df.isnull().sum().sum())
print("Data cleaning complete!")
```

## 6. Creating new column Country by extracting it from Location column

```python
df['Country'] = df['Locations'].astype(str).str.split(', ').str[-1]
df.loc[df['Locations'] == 'Unknown', 'Country'] = 'Unknown'
print("Created 'Country' column from 'Locations'.")
display(df[['Locations', 'Country']].head())
```

## 7. Data Visualization 

```python
plt.figure(figsize=(12, 8))
top_10_countries = df['Country'].value_counts().nlargest(10)
sns.barplot(x=top_10_countries.values, y=top_10_countries.index, palette='crest')
plt.title('Top 10 Countries by Number of Clinical Trials', fontsize=16)
plt.xlabel('Number of Trials', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Status', data=df, order=df['Status'].value_counts().index, palette='flare')
plt.title('Distribution of Clinical Trial Status', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Status', fontsize=12)
plt.show()
```

```python
gender_counts = df[df['Gender'] != 'Unknown']['Gender'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',
        startangle=140, colors=sns.color_palette('pastel'))
plt.title('Gender Distribution in Clinical Trials', fontsize=16)
plt.show()
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
df['Start Month'] = df['Start Date'].dt.month_name()
plt.figure(figsize=(12, 7))
sns.countplot(x='Start Month', data=df, order=df['Start Month'].value_counts().index, palette='viridis')
plt.title('Number of Trials Started by Month', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Trials', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
top_10_countries_list = df['Country'].value_counts().nlargest(10).index
df_top_countries = df[df['Country'].isin(top_10_countries_list)]
status_country_crosstab = pd.crosstab(df_top_countries['Status'], df_top_countries['Country'])
plt.figure(figsize=(14, 10))
sns.heatmap(status_country_crosstab, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Distribution of Trial Status Across Top 10 Countries', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Trial Status', fontsize=12)
plt.show()
```

```python
df_sorted_by_date = df.sort_values(by='Start Date').dropna(subset=['Start Date'])
df_sorted_by_date['Cumulative Trials'] = range(1, len(df_sorted_by_date) + 1)
plt.figure(figsize=(12, 7))
plt.plot(df_sorted_by_date['Start Date'], df_sorted_by_date['Cumulative Trials'], marker='.', linestyle='-', markersize=2)
plt.title('Cumulative Growth of COVID-19 Clinical Trials Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Number of Trials (Cumulative)', fontsize=12)
plt.grid(True)
plt.show()
```
